```python
import numpy as np
import pandas as pd
import random 
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

# Problem 1: Plot a rugplot on given below random generated data.
num = np.random.randn(170)

```python
num = np.random.randn(170)
sns.distplot(num,hist=False,kde=False,rug=True)
plt.show()
```

# Problem 2: Plot a pairs plot on iris data set showing different species. Write your conclusions as well.

```python
iris=sns.load_dataset('iris')
sns.pairplot(iris,hue='species',kind='reg',corner=True)
plt.show()
```

Observation:
1. Petal_length can be used for analysis as there is least overlapping
2. Petal_width and Petal_length have a linear relation(roughly)

# Problem 3: Plot a joint plot showing relation between 'petal length' and 'petal width' with density estimates.

```python
sns.jointplot(x='petal_length',y='petal_width',kind='kde',data=iris)
plt.show()
```

# Problem 4: Plot a pie chart of survived based on sex.(on Titanic Dataset)

```python
titanic=sns.load_dataset('titanic')
data=titanic[titanic['survived']==1]
labels=data.groupby('sex')['sex'].count().index.to_list()
vals=data.groupby('sex')['sex'].count().to_list()
```

```python
plt.pie(vals,labels=labels,autopct='%0.2f')
plt.show()
```
